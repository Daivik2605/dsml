```python
# sort the indexing and rename 'id' in cust_df
cust_df.rename(columns = {'id': 'cust_id'}, inplace = True)
```

```python
# Lets merge the dataframes, inner join of cust_df and tran_df\
```

```python
merged_df = pd.merge(
    left = tran_df,
    right = cust_df2,
    left_on = 'customer_id',
    right_index = True
)
```

### For each customer, what are the minimum, max, and average amounts_spent_per_transaction

```python
# Lets use aggregate
merged_df.groupby('cust_name') ['txn_total'].aggregate(['min', 'max', 'mean'])
```

```python
# another way to use aggregate
merged_df.groupby('cust_name').aggregate({'txn_total': ['min', 'max', 'mean']})
```

### For each banking_category, find the total amount spent:

```python
merged_df.groupby('cust_category')['txn_total'].sum()
```

```python
merged_df.groupby('cust_category')['txn_total'].agg(['sum'])
```
